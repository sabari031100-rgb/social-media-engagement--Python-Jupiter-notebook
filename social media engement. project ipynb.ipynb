```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Total reactions & engagement calculation
reaction_cols = ['num_likes', 'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']
df['total_reactions'] = df[reaction_cols].sum(axis=1)
df['total_engagement'] = df['total_reactions'] + df['num_comments'] + df['num_shares']
```

```python
status_type_cols = ['status_type_link', 'status_type_photo', 'status_type_status', 'status_type_video']
avg_engagement = {}
for col in status_type_cols:
    avg_engagement[col.replace("status_type_", "")] = df.loc[df[col] == 1, 'total_engagement'].mean()

plt.figure(figsize=(8,5))
sns.barplot(x=list(avg_engagement.keys()), y=list(avg_engagement.values()), palette="viridis")
plt.title("📊 Average Engagement by Post Type", fontsize=14)
plt.ylabel("Average Engagement")
plt.xlabel("Post Type")
plt.show()
```

```python
reaction_sum = df[reaction_cols].sum()
plt.figure(figsize=(7,7))
plt.pie(reaction_sum, labels=reaction_sum.index, autopct='%1.1f%%', startangle=140)
plt.title("🍰 Reaction Type Distribution", fontsize=14)
plt.show()
```

```python
avg_likes_photo = df.loc[df["status_type_photo"] == 1, "num_likes"].mean()
avg_likes_video = df.loc[df["status_type_video"] == 1, "num_likes"].mean()
avg_likes_link  = df.loc[df["status_type_link"]  == 1, "num_likes"].mean()
avg_likes_status= df.loc[df["status_type_status"]== 1, "num_likes"].mean()

plt.bar(["Photo", "Video", "Link", "Status"], [avg_likes_photo, avg_likes_video, avg_likes_link, avg_likes_status], color=["blue","green","orange","red"])
plt.title("👍 Average Likes by Post Type")
plt.ylabel("Average Likes")
plt.show()
```

```python
avg_comments_photo = df.loc[df["status_type_photo"] == 1, "num_comments"].mean()
avg_comments_video = df.loc[df["status_type_video"] == 1, "num_comments"].mean()
avg_comments_link  = df.loc[df["status_type_link"]  == 1, "num_comments"].mean()
avg_comments_status= df.loc[df["status_type_status"]== 1, "num_comments"].mean()

plt.bar(["Photo", "Video", "Link", "Status"], 
        [avg_comments_photo, avg_comments_video, avg_comments_link, avg_comments_status],
        color=["blue","green","orange","red"])
plt.title("💬 Average Comments by Post Type")
plt.ylabel("Average Comments")
plt.show()
```
